```python
import pdb
Heads = [] 
inputfile = open('PGPgiantcompo.graph')
#inputfile = open('test1.txt')
i = 0
for line in inputfile:
    Heads.append(line.strip())
    Heads[i] = Heads[i].split()
    for j in range(len(Heads[i])):
        #print(type(Heads[i][j]))
        Heads[i][j] = int(Heads[i][j])
    i = i + 1
Heads = Heads[1:]
#print((Heads[2]))
```

```python
Tails = range(1, len(Heads)+1) 
```

```python
Adj_list = {} 

for i in range(len(Tails)):
    Adj_list[Tails[i]] = Heads[i]
#print(Adj_list)
```

```python
max_root_dict = {} 
Q = [] 
for root_vertex in Tails:
    dist_dict = {}
    dist = [100 for x in range(len(Tails))]
    #print(len(Tails))
    for i in range(len(Tails)):
        dist_dict[Tails[i]] = dist[i]
    #Q.append(root_vertex)
    #root_vertex = 8
    Q.append(root_vertex)
#dist_dict[root_vertex] = 0
    dist_dict[root_vertex] = 0
    while Q:
        v = Q.pop(0)
        #pdb.set_trace()
        for t in Adj_list[v]:
            #print(type(w))
            #print(dist_dict[w])
            if dist_dict[t] == 100:
                dist_dict[t] = dist_dict[v] + 1   
                Q.append(t)
    max_root_dict[root_vertex] = max(list(dist_dict.values()))
    #print(max(list(dist_dict.values())))
    Q = []
print(max_root_dict)
print('Diameter of the graph is:' , max(list(max_root_dict.values())))
```
